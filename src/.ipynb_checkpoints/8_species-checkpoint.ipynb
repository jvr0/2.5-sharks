# Limpieza columna "Species"

```python
df.Species = df.Species.apply(lambda x: str(x))
df.Species.str.lower()
```

```python
def especies(x):
    if "whit" in x:
        return "White Shark"
    elif "hamm" in x:
        return "Hammerhead Shark"
    elif "Tige" in x:
        return "Tiger Shark"
    elif "bull" in x:
        return "Bull Shark"
    elif "lemo" in x:
        return "Lemon Shark"
    elif "mako" in x :
        return "Mako Shark"
    elif "nurs" in x:
        return "Nurse Shark"
    elif "blac" in x:
        return "Blacktip Shark"
    elif "ocea" in x:
        return "Oceanic Whitetip Shark"
    elif "thre" in x:
        return "Thresher shark"
    elif "silk" in x:
        return "Silky Shark"
    elif "dusk" in x:
        return "Dusky shark"
    elif "gala" in x:
        return "Galapagos Shark"
    elif "gobl" in x:
        return "Goblin Shark"
    elif "ange" in x:
        return "Angel Shark"
    elif "gree" in x:
        return "Greenland Shark"
    elif "carp" in x:
        return "Carpet Shark"
    elif "shark" in x:
        return "Shark"
    else:
        return x
```

# Creación columna "Size"

```python
def tamaño(x):
    try:
        numeros = re.findall(r'\d+', x)
        return [int(numero) for numero in numeros]
    except:
        return x
```

```python
df.Size = df.Species.apply(tamaño)
```

```python
df.Size = df.Size.apply(lambda x: round(np.mean(x),2))
df.Size = df.Size.apply(lambda x: np.nan if x > 100 else x)
df.Size = df.Size.fillna(value = df.Size.mean())
df.Size = df.Size.apply(lambda x: round(x,2))
```
